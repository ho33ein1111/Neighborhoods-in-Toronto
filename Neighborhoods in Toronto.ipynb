```python
import numpy as np # library to handle data in a vectorized manner
import pandas as pd # library for data analsysis
import json # library to handle JSON files
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe
import matplotlib.pyplot as plt
%matplotlib inline
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests # library to handle requests
import csv
```

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html = urlopen(url)
source = requests.get(url)
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

#### Finding html code related to table in webpage

```python
table = soup.find( 'table' , class_ = 'wikitable sortable')
print(table.prettify())
type(table)
```

#### using panda read_html to create a DataFrame 

```python
df = pd.read_html(str(table))
df_table = df[0]
df_table
```

### Cleaning Data 

```python
#Ignore cells with a borough that is Not assigned
df_table = df_table[df_table['Borough'] != 'Not assigned' ]

```

##### if a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough

```python
df_table[df_table['Neighbourhood'] == 'Not assigned'  ]
```

```python
df_table.loc[df_table['Neighbourhood']=='Not assigned' , 'Neighbourhood' ] = df_table[df_table['Neighbourhood'] == 'Not assigned' ]['Borough']
```

#### Group df by Postcode and join other column by ','

```python
df_group = df_table.groupby('Postcode').agg(lambda x: ','.join(x.unique()) ).reset_index()

df_group
```
